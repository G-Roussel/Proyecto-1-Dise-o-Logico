    empieza un contador tipo bandera donde se le asigna a una variables entero, usando dos banderas y restandolas obtenemos el tiempo de ejecución del algoritmo

El programa se divide principalmente en 4 etapas diferenciadas, donde se segmentan en funciones debidamente descritas con comentarios dentro del codigo. En general, las etapas se describen a continuación:

    En la primera etapa convertimos los numeros a binarios y guardamos en una lista de string
    En la segunda etapa agrupamos los numeros binarios según su indice, es decir, cuantos 1 tienen en su representacion binaria y lo guardamos en un 
    En la tercer etapa se obtienen los implicantes primos generales para luego clasificarlos 
    En la cuarta etapa se obtienen los implicantes primos escenciales que corresponden a la ecuacion booleana reducida

Adicionalmente se implementa una interfaz para usuario

```python
import time
import tkinter
import tkinter.messagebox
ventana=tkinter.Tk()
ventana.geometry("400x300")
ventana.title('DISEÑO LOGICO PROYECTO 1')

etiqueta=tkinter.Label(ventana, text=" Bienvenidos",font="helvetica",background="green")
etiqueta.pack(side=tkinter.TOP)
etiqueta.pack(fill=tkinter.X)
etiqueta=tkinter.Label(ventana, text=" Algoritmo de Quine-McCluskey",font="helvetica",background="green")
etiqueta.pack(side=tkinter.BOTTOM)
etiqueta.pack(fill=tkinter.X)

def ventana_algortimo():
    ventana = tkinter.Toplevel()
    ventana.geometry("400x300")
    ventana.title('DISEÑO LOGICO PROYECTO 1')
    # variable para el numero de literales y nos aseguramos que sea un entero
    label = tkinter.Entry(ventana)
    label.pack()
    label2 = tkinter.Entry(ventana)
    label2.pack()
    print("ingrese el numero de literales, ")
    opcion = tkinter.Button(ventana,text="Ingrese primero el numero de literales y, \n"
                                         " luego los minterminos separados por espacio,\n"
                                         "y dale a la opcion ejecutar algoritmo",padx=78, pady=30, background="blue")
    opcion.pack()

    opcion = tkinter.Button(ventana, text="EJECUTAR ALGORTIMO", padx=60, pady=30, background="blue", command=lambda: algoritmo(label.get(),label2.get()))
    opcion.pack()
    opcion = tkinter.Button(ventana, text="SALIR", padx=78, pady=30, background="blue", command=salir)
    opcion.pack()

#Establecemos la función principal, donde se estructuran las demas funciones que se encargan de filtrar los minterminos.
#Primero que todo necesitamos el tamaño de la matriz donde se alojarán los minterminos para su clasificación
    
def algoritmo(r,minterminos):
    print("ingrese el numero de literales")
    numero_literales = int(r)
    print("seleccionó", numero_literales, "literales")
    # Donde se alojan los minterminos
    print("ingrese los minterminos separados por espacio")

    minterminos = minterminos.split()
    minterminos = [int(x) for x in minterminos]
    print("los minterminos son", minterminos)

    # nos aseguramos que el numero de variables coincida con la cantidad de minterminos ingresados
    if len(minterminos) == numero_literales:
        print("Iniciando calculos")
        start = time.perf_counter()
        # Primera etapa
        numeros_binarios = convertir_binario(numero_literales, minterminos)
        # Segunda etapa
        segunda = agrupacion(numeros_binarios)
        print("segunda etapa:Implicantes primos son ", segunda)

        tabla = tabla_de_primos(segunda, numeros_binarios)

        escenciales = primos_escenciales(tabla, segunda)

        print("Implicantes primos escenciales:", escenciales)

    else:
        print("error en la digitación de parametros, ingrese nuevamente")
    end=time.perf_counter()
    print("el tiempo de ejecucion es: ")
    print( end- start)## tiempo de ejecucion del algoritmo

#####
# Primera etapa, convertir los numeros a binario, recibe un entero y una lista y retorna una lista
def convertir_binario(numero_literal, minterminos):
    # creamos una lista donde se alojarán los binarios
    binarios = []

    # creamos un string donde se alojaran los terminos antes de ser ordenados
    a = ""
    for i in minterminos:
        for j in range(numero_literal):
            a = str(i % 2) + a
            i //= 2
        binarios.append(a)
        a = ""
    print("la conversión es", binarios)
    return binarios


#####
# Segunda etapa, agrupamos los numeros binarios según su indice, es decir, cuantos 1 tienen en su representacion binaria
def agrupacion(binarios):
    # creamos una lista vacía para luego retornar los resultados del algoritmo
    ret = []
    # iniciamos un bucle para comparar cada binario y agruparlos
    while 1:
        marca = ["bin"] * len(binarios)
        # lista temporal
        t = []
        # con este for recorremos la lista de binarios y comparamos diferentes posiciones
        for i in range(len(binarios)):
            for j in range(i + 1, len(binarios)):
                # determinamos una bandera al comparar los binarios
                x = 0
                # convertimos a lista cada numero binario
                lista1 = list(binarios[i])
                lista2 = list(binarios[j])
                # añadimos un contador que nos indica si cada posición de la lista es diferente
                contador = 0
                for k in range(len(lista1)):
                    if lista1[k] != lista2[k]:
                        # si las posiciones son diferentes
                        contador += 1
                        lista1[k] = "_"

                if contador > 1:
                    x = -1
                if contador <= 1:
                    x = "".join(lista1)

                if x != -1:
                    marca[i] = "*"
                    marca[j] = "*"
                    t.append(x)
        for i in range(len(binarios)):
            if marca[i] == "bin":
                ret.append(binarios[i])
        if len(t) == 0:
            return ret
        binarios = list(set(t))
    print()


#####
# Tercera etapa,
def tabla_de_primos(segunda, numeros_binarios): ## se crea una tabla de los primos
    tabla = [[0 for x in range(len(numeros_binarios))] for x in range(len(segunda))]
    for i in range(len(segunda)):
        contador = segunda[i].count("_")
        for j in range(len(numeros_binarios)):
            if encontrar_escenciales(segunda[i], numeros_binarios[j], contador):
                tabla[i][j] = 1

    return tabla


####
# Cuarta etapa
def primos_escenciales(tabla, segunda):## en esta funcion se determinan los valores de los primos escenciales
    lista_auxiliar = []
    elije = [0] * len(tabla)
    for i in range(len(tabla[0])):
        contador = 0
        aux = -1
        for j in range(len(tabla)):
            if tabla[j][i] == 1:
                contador += 1
                aux = j
        if contador == 1:
            elije[aux] = 1
    for i in range(len(elije)):
        if elije[i] == 1:
            for j in range(len(tabla[0])):
                if tabla[i][j] == 1:
                    for k in range(len(tabla)):
                        tabla[k][j] = 0
            lista_auxiliar.append(segunda[i])
    while 1:
        maximo = 0
        aux = -1
        cont_aux = 0
        for i in range(len(tabla)):
            cont_aux = tabla[i].count(1)
            if cont_aux > maximo:
                maximo = cont_aux
                aux = i

        if maximo == 0:
            return lista_auxiliar

        lista_auxiliar.append(segunda[aux])

        for i in range(len(tabla[0])):
            if tabla[aux][i] == 1:
                for j in range(len(tabla)):
                    tabla[j][i] = 0


#####
# Funciones auxiliares, para econtrar los primos escenciales
def encontrar_escenciales(segunda, numeros_binarios, contador):
    lista1 = list(segunda)
    lista2 = list(numeros_binarios)
    cont_aux = 0
    for i in range(len(lista1)):
        if lista1[i] != lista2[i]:
            cont_aux += 1
    if cont_aux == contador:
        return True
    else:
        return False



## funcion para informacion
def info():
    print("ESTUDIANTES")
    ventana = tkinter.Toplevel()
    ventana.geometry("400x300")
    ventana.title('INFORMACIÓN')
    opcion = tkinter.Button(ventana, text="David Pérez y George Rousell, proyecto 1 de diseño logico", padx=78, pady=30, background="white")
    opcion.pack()
    opcion = tkinter.Button(ventana, text="SALIR", padx=78, pady=30, background="blue", command=salir)
    opcion.pack()

##funcion para salir del programa
def salir():
    tkinter.messagebox.showinfo("Salir","!Gracias!")
    ventana.destroy()

## creando las opciones que digita el usuario para recorrer el algortimo
opcion= tkinter.Button(ventana, text="ALGORTIMO", padx=60, pady=30, background="blue", command=ventana_algortimo)
opcion.pack()
etiqueta.pack(fill=tkinter.X)

opcion= tkinter.Button(ventana, text="INFO", padx=80, pady=30, background="blue", command=info)
opcion.pack()
etiqueta.pack(fill=tkinter.X)
opcion= tkinter.Button(ventana, text="SALIR", padx=78, pady=30, background="blue", command=salir)
opcion.pack()
etiqueta.pack(fill=tkinter.X)


ventana.mainloop()

```

Algoritmo Espresso
    
    Tenemos los métrodos clasicos de minimización de funciones booleanas como los mapas de Karnaugh. Este puede realizarse a mano, aunque es un proceso laborioso y tardado. Por otro lado el primer método alternativo desarrollado por Willar Quine y Edward McCluskey que simplificó la manera de resolver estas funciones y mas con la ejecución e implementación en programas de computadora.
    El algotirmo espresso, desarrollado por Brayton et al en la universidad de California, Berkeley, manipula cubos que son minterminos del producto en las cubiertas ON, DC y OFF iterativamente, resultanto en una aproximación muy cercana al mínimo global, pero con un uso mas eficiente del tiempo de ejecución y de la memoria.
    Es muy util para matrices de puertas programables en campo o un circuito integrado específico de la aplicación
